```python
import matplotlib.pyplot as plt
```

```python
# create a bar chart f0r ship mode
count_ship_mode = cat_cols['Ship_Mode'].value_counts()

#for the shape and size
plt.figure(figsize = (8,6))

#for customisation
count_ship_mode.plot(kind = 'bar', color = 'brown' ,edgecolor = 'purple')
    
plt.title('Count of ship mode')
plt.ylabel('Ship_Mode')
plt.xlabel('Count of ship mode')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
           

           
```

This is a univariate analysis becasuse we are working only on one column at a time

From the analysis, the company operates on 4 different modes of shipping which are the standard class, second class, first class and same day. The standard class mode of shipping has the highest count followed by the second class with the same day shippin mode being the lowest. From this analysis ,we can conclude that customers prefer to use the standard class shipping mode compared to other shipping mode.The reason for this is because the standard class shipping mode offers a cheaper rate and also compared to the other shipping mode

```python
# sale segment by sales
sales_segment = df1.groupby('Segment')['Sales'].sum().reset_index()

# figure of bar chart

#sales_segment.plot(kind = 'bar', color = 'skyblue')

plt.bar(sales_segment['Segment'],sales_segment ['Sales'],color = 'skyblue')
plt.title('Total sales by Segment')
plt.ylabel('Total sales')
plt.xlabel('Segment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# sales by region
df1['Region'].unique()
```
