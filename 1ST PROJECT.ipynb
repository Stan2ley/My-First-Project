```python
df = df.drop(columns = ['Occupation Type' ,'Cnt Children'] , axis = 1)
```

```python
df.columns = (df.columns.str.replace("Cnt", "Count") ) 
```

```python
df['Flag Phone'] = df['Flag Phone'].replace({'Y': 'Yes','N': 'No'})
```

```python
df['Code Gender'] = df['Code Gender'].replace({ 'M': 'Male','F': 'Female'})
```

```python
flag_columns = [
    'Flag Mobil',
    'Flag Work Phone',
    'Flag Email',
    'Flag Phone']
df[flag_columns] = df[flag_columns].replace({1: 'Yes',0: 'No'})
```

```python
flag_columns = [
    'Flag Own Realty',
    'Flag Own Car']
df[flag_columns] = df[flag_columns].replace({'Y': 'Yes', 'N': 'No'})
```

```python
df['Days Employed'] = (abs(df['Days Employed']) / 365).astype(int)
```
